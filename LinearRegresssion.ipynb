```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_train = pd.read_csv("Train.csv")
data_test= pd.read_csv("Test.csv")
```

```python
data_train['Fuel_Type'] = data_train['Fuel_Type'].replace({'Petrol': 1, 'Diesel': 0, 'Electric': 3})
data_train['Transmission'] = data_train['Transmission'].replace({'Manual': 1, 'Automatic': 0})
data_train.head()
data_test['Fuel_Type'] = data_test['Fuel_Type'].replace({'Petrol': 1, 'Diesel': 0, 'Electric': 3})
data_test['Transmission'] = data_test['Transmission'].replace({'Manual': 1, 'Automatic': 0})
```

```python
def extract(value):
    try:
        return float(value.split()[0])
    except:
        return np.nan
```

```python
for i in ['Mileage','Engine', 'Power']:
    data_train[i] = data_train[i].apply(extract)
    data_test[i] = data_test[i].apply(extract)
data_train = data_train.dropna()
data_test= data_test.dropna()
```

```python
updatedlist_train=pd.concat([city,owner_type],axis='columns')
updatedlist_train=updatedlist_train.astype(int)
updatedlist_test=pd.concat([city_d2,owner_type_d2],axis='columns')
updatedlist_test=updatedlist_test.astype(int)
```

```python
newdata_train = pd.concat([data_train,updatedlist_train], axis=1)
newdata_train.head()
newdata_test = pd.concat([data_test,updatedlist_test], axis=1)
```

```python
newdata_train = newdata_train.drop(['Unnamed: 0','Name','Location','Owner_Type'],axis = 1)
# newdata_train
newdata_test = newdata_test.drop(['Unnamed: 0','Name','Location','Owner_Type'],axis = 1)
newdata_test
```

```python
X_train = newdata_train
y_train = X_train['Price']
X_train = X_train.drop(['Price'],axis = 1)
y_train = y_train.dropna()

#for test.csv
X_test = newdata_test
y_test = X_test['Price']
X_test = X_test.drop(['Price'],axis = 1)
y_test = y_test.dropna()
```

```python
X_train = (X_train - X_train.mean()) / X_train.std()
y_train = (y_train - y_train.mean()) / y_train.std()
X_test= (X_test - X_test.mean()) / X_test.std()
y_test = (y_test - y_test.mean()) / y_test.std()
```

```python
# Calculate R-squared.
def r2_score(y_true, y_pred):
   
    total_variance = np.sum((y_true - np.mean(y_true)) ** 2)
    explained_variance = np.sum((y_pred - y_true) ** 2)
    r2 = 1 - (explained_variance / total_variance)
    return r2 * 100  # to express as a percentage
```

```python
  # Calculate RMSE cost.
def RMSE(y, y_pred):
    return np.sqrt(np.mean((y_pred - y) ** 2))
```

```python
def gradient_descent(X_train, bias, alpha, iterations, y_train):
    X = np.array(X_train)
    y = np.array(y_train)
    
    samples, columns = X.shape
    weights = np.zeros(columns)  # Initialize weights as zeroes
    y_train_pred = []
    y_pred=0
    ycost=[]
    for k in range(iterations):
        y_pred = np.dot(X, weights) + bias 

        # Calculate gradients
        error = y_pred - y  
        dw = (1 / samples) * np.dot(X.T, error)
        db = (1 / samples) * np.sum(error)

        # Update weights and bias
        weights -= alpha * dw
        bias -= alpha * db

        # Compute and store cost for monitoring
        cost = RMSE(y, y_pred)
        y_train_pred.append(cost)
        ycost.append(y_pred)

        # Print epoch details and accuracy
        print("Epoch:", k + 1, "cost:",cost, "Accuracy:", r2_score(y, y_pred), "%")

    return weights, bias, y_train_pred,ycost
```

```python
w, b, yaxis, ycost = gradient_descent(X_train, 0, 0.6, 200, y_train)
print(w)
print(b)
```

```python
def result(x,y,weights,bias):
    samples, columns = x.shape
    cost = 0
    ypredict = []
    for j in range(samples):
        ypred_test = 0
        for i in range(columns):
            ypred_test += x.iloc[j,i]*weights[i]+bias
        cost += (ypred_test-y.iloc[j])**(2)
        ypredict.append(ypred_test)
    cost = (cost/samples)**(0.5)
    return cost, r2_score(y_test,ypredict)
```

```python
Cost_test, Accuracy_test = result(X_test,y_test,w,b)
```

```python
xaxis = list(range(1, len(yaxis)+1))
plt.plot(xaxis,yaxis)
plt.xlabel("Epochs")
plt.ylabel("Cost")
plt.show()
```
